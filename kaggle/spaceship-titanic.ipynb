```python
train['Transported'].replace({False: 0, True: 1}, inplace=True)
```

```python
train['CryoSleep'] =train['CryoSleep'].fillna('False')
train['VIP'] =train['VIP'].fillna('False')
train['HomePlanet'] =train['HomePlanet'].fillna('Earth')
train['Destination'] =train['Destination'].fillna('TRAPPIST-1e')
train['Age'] =train['Age'].fillna(train['Age'].mean())
```

```python
train['CryoSleep'].replace({False: 0, True: 1},inplace=True)
train['VIP'].replace({False: 0, True: 1},inplace=True)
train['HomePlanet'].replace(('Earth','Europa','Mars'),(1,2,3),inplace=True)
train['Destination'].replace(('TRAPPIST-1e','PSO J318.5-22','55 Cancri e'),(1,2,3),inplace=True)
```

```python
train.drop(["Cabin","PassengerId","RoomService","FoodCourt","ShoppingMall","Spa","VRDeck","Name"],axis=1,inplace=True)
test.drop(["Cabin","PassengerId","RoomService","FoodCourt","ShoppingMall","Spa","VRDeck","Name"],axis=1,inplace=True)
```

```python
X = train.drop(columns=('Transported'))
y = train['Transported']

print("Feature values: ", X)
print("Labels: ", y)
```

```python
print(X.dtypes)
```

```python
X.astype(int)
```

```python
from sklearn import preprocessing

X = preprocessing.MinMaxScaler().fit_transform(X)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print('Train set:', X_train.shape,  y_train.shape)
print('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
```

```python
from sklearn import metrics

prediction = knn.predict(X_test)

print("Train set Accuracy: ", metrics.accuracy_score(y_train, knn.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, prediction))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(16, input_shape=(5,), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1))
          
model.compile(optimizer=Adam(learning_rate=0.005), loss='binary_crossentropy',metrics=['Accuracy'])

model.summary()
```

```python
train_hist = model.fit(X_train, y_train, validation_data = (X_test, y_test), verbose=1, epochs=10)
```
