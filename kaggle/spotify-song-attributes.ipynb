```python
df = pd.read_csv('/kaggle/input/spotifyclassification/data.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head(10)
```

```python
df.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['artist'] = encoder.fit_transform(df['artist'])
df.head(5)
```

```python
correlation_matrix = df.corr()
print(correlation_matrix.target.sort_values(ascending=False))
```

```python
from sklearn.preprocessing import StandardScaler

features = ['danceability', 'speechiness', 'instrumentalness', 'duration_ms', 'valence']

X = df[features].values
y = df['target'].values

X = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test ,y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)

print('Train set accuracy: ', accuracy_score(y_train, model.predict(X_train)))
print('Test set accuracy: ', accuracy_score(y_test, pred))
```

```python
from sklearn.svm import SVC

model = SVC().fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)

print('Train set accuracy: ', accuracy_score(y_train, model.predict(X_train)))
print('Test set accuracy: ', accuracy_score(y_test, pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4).fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)

print('Train set accuracy: ', accuracy_score(y_train, model.predict(X_train)))
print('Test set accuracy: ', accuracy_score(y_test, pred))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)
```
